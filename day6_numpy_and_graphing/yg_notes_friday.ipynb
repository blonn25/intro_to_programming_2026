#### Use matplotlib for data visualization 

```python
x = np.linspace(-5, 5, 50) #generate 50 data points between -5 and 5 
y = x ** 2 
plt.plot(x,y, color = 'green', label = 'y = x^2')  
plt.scatter(x, y, color = 'purple') 

#label our plots 
plt.xlabel("x")
plt.ylabel("y")
plt.title("Exponential Function")
plt.legend()
#save plot in your current directory 
# plt.savefig('exponential_function.png', dpi = 600)
```

```python
time = np.arange(0, 10, 0.1) #step size 0.1, from 0 to 10, 1d array of 100 
sine_wave = np.sin(time) #1d array of 100
noisy_wave = sine_wave + np.random.normal(0, 0.1, len(sine_wave)) #1d array of 100 

#plot sin wave 
plt.plot(time, sine_wave, label = 'Sine Wave', color = 'blue')
#plot noisy sin wave 
plt.plot(time, noisy_wave, label = 'Noisy Wave', color = 'orange')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.legend() #shows labels on the graph
plt.show()
```

```python
plt.hist(noisy_wave, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Noisy Sine Wave')
# plt.show()
```
